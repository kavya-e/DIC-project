#### filling null values with mode and removing the outlier data of 1% group size.

```python


#----------------------------------------->>>>>>>>>>>     2      <<<<<<<<<<---------------------------------------------------
```

```python
grouped = traffic_crashes.groupby('TRAFFICWAY_TYPE')
Empty_list_group = []
# Define a function to remove rare values within each group
def remove_rare_values(group , groupkey):
    # Replace NaN values in 'LANE_CNT' with the mode of the group
    mode_values = group['LANE_CNT'].mode()
    if not mode_values.empty:  # Check if mode calculation produced any values
            mode = mode_values.iloc[0]  # Access the first mode value
            group['LANE_CNT'].fillna(mode, inplace=True)  # Replace NaN values with mode
    else:
        Empty_list_group.append(groupkey)
    # Count the frequency of each unique value in the group
    value_counts = group['LANE_CNT'].value_counts()
    # Define a threshold for identifying rare values (e.g., 1% of the group size)
    threshold = 0.01 * len(group)
    # Identify rare values with frequencies below the threshold
    rare_values = value_counts[value_counts < threshold].index
    # Remove rows where 'LANE_CNT' corresponds to rare values
    return group[~group['LANE_CNT'].isin(rare_values)]

# Apply the function to each group and concatenate the results
df_filtered = pd.concat([remove_rare_values(group , groupkey) for groupkey, group in grouped])

# Reset index if necessary
df_filtered.reset_index(drop=True, inplace=True)

df_filtered.info()

```

# K Means

```python
import numpy as np
from scipy.stats import mode
from sklearn.cluster import KMeans

# Initialize the K-Means model
kmeans = KMeans(n_clusters=3, random_state=5)  #The cluster size is equal to 3
# Fit the K-means model
kmeans.fit(X_train_scaled)

# Predict cluster labels for the test set
kmeans_pred = kmeans.predict(X_test_scaled)

#For calculating the accuracy
# Assigning class labels to clusters based on the mode of true labels in each cluster
cluster_labels = np.zeros_like(kmeans_pred)
for i in range(3):  # Since there are 3 clusters
    mask = (kmeans_pred == i)
    cluster_labels[mask] = mode(y_test[mask])[0]

accuracy = np.sum(cluster_labels == y_test) / len(y_test)
print(f"Accuracy of K-Means clustering: {accuracy * 100:.2f}%")


```

```python
import gradio as gr
import pickle
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import StandardScaler

# Assuming X_train and y_train are defined, and X_train is a DataFrame

# Specify all relevant feature names
relevant_feature_names = ['CRASH_HOUR','CRASH_DAY_OF_WEEK', 'Weather_Weight', 'Lighting_Weight', 'road_defect_weights']

# Get the column indices corresponding to the relevant feature names
relevant_feature_indices = [X_train.columns.get_loc(name) for name in relevant_feature_names]

# Select only the relevant features used for prediction
X_train_relevant = X_train.iloc[:, relevant_feature_indices]

# Scale the relevant features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train_relevant)

# Fit the decision tree classifier
clf = DecisionTreeClassifier(random_state=42)
clf.fit(X_train_scaled, y_train)

# Save the model after training
with open("model.pkl", "wb") as f:
    pickle.dump((clf, scaler), f)  # Save both the model and scaler


def make_prediction(crash_hour, CRASH_DAY_OF_WEEK, weather_weight, lighting_weight, road_defect_weights):
    # Load the trained classifier and scaler from the pickle file
    with open("model.pkl", "rb") as f:
        clf, scaler = pickle.load(f)

    # Prepare the input features as expected by the model
    features = [[crash_hour, CRASH_DAY_OF_WEEK, weather_weight, lighting_weight, road_defect_weights,]]
    # Scale the input features
    scaled_features = scaler.transform(features)
    preds = clf.predict(scaled_features)

    # Determine the damage category based on the prediction
    if preds[0] == 1:
        result = "Damage will be Less than 500$"
    elif preds[0] == 2:
        result = "Damage is in between 500$ to 1500$"
    else:
        result = "Damage is over $1500"
    return result

# Create Gradio inputs for each feature
crash_hour_input = gr.Number(label="Enter the first car crash hour {0 to 12} ")
crash_day_of_week_input = gr.Number(label="Enter the car crash day of the week (1: Monday, 7: Sunday)")
weather_weight_input = gr.Number(label="Enter the weather (1: CLEAR, 2: CLOUDY, 0: UNKNOWN)")
lighting_weight_input = gr.Number(label="Enter the lighting condition (0: UNKNOWN, 1: DAYLIGHT, 2: DARKNESS)")
road_defect_weights_input = gr.Number(label="Enter the road defect type (1: NO DEFECTS, 2: DEFECTIVE, 0: UNKNOWN)")


# Gradio interface
app = gr.Interface(fn=make_prediction, inputs=[crash_hour_input, crash_day_of_week_input,weather_weight_input, lighting_weight_input, road_defect_weights_input], outputs="text")
app.launch()
```
